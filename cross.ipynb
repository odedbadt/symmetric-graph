```python
import numpy as np
```

```python
%matplotlib inline

import matplotlib
from matplotlib import pyplot as plt
```

```python
def normalize(V):
    norm = np.linalg.norm(V)
    return V / norm if norm > 0 else V
```

```python
V = np.random.randn(4, 3)
W = np.zeros((1, 3))
for i in range(3):
    n = (i + 1) % 3
    W = W + normalize(np.cross(V[i + 1, :] - V[0, :], V[n + 1, :] - V[0, :]))

M = np.concatenate([V, W], axis=0)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
p = [0, 1, 0, 2, 0, 3]
p2 = [0, 4]
plt.plot(M[p,0], M[p,1], M[p,2], 'r')
plt.plot(M[p2,0], M[p2,1], M[p2,2], 'b')
plt.show()
```
