```python
%matplotlib notebook
import matplotlib
from matplotlib import pyplot as plt
```

```python
import numpy as np
(S, V, D) = np.linalg.svd(np.random.normal(size=(3, 3)))
```

```python
def normalize(v):
    return v / np.linalg.norm(v)
```

```python
n = 5
```

```python
W = np.zeros((3, n+1), 'float')
```

```python
np.cos(np.pi)
```

```python
W[0, 0] = 1
for i in range(n):
    alpha = i * 2.0 * np.pi / n
    W[1, i + 1] = np.cos(alpha)
    W[2, i + 1] = -np.sin(alpha)
O = np.matmul(S, W)
```

```python
def draw_line(V1, V2, style):
    plt.plot([V1[0], V2[0]], [V1[1], V2[1]], [V1[2], V2[2]], style)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
origin = [0, 0, 0]
F = normalize(np.random.normal(size=(3, 1)))
for i in range(n + 1):
     draw_line(origin, O[:, i], 'b' if i == 0 else 'r')
draw_line(origin, O[:, 1], 'g')
draw_line(origin, F, 'k')
C = np.cross(F.flatten(), O[:,1])
draw_line(origin, C, 'm')
```
